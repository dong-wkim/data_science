## Exploratory Data Analysis (EDA) Tasks  

This is a checklist of all the EDA tasks required for this project.
While you may explore additional tasks beyond this list, you must
complete all the basic requirements outlined here to meet the project
expectations. 

### Describe the Variable Characteristics with Summary Statistics  

Provide description of the variables you are using in this project.  

### Conduct Univariate Analysis 

Calculate descriptive statistics (mean, median, mode, variance,
skewness, kurtosis) of the key variables in your dataset 

Create visualization of the key variables to examine distribution or
patterns 

Consider adding histogram, box plots, or density plots to visualize the
distribution of numerical variables, and bar charts for categorical
variables. 

If you have to handle outliers/skewness, it is important to visualize
outliers through boxplots or scatterplots and then explain the methods
used, such as why trimming or transformation was applied (or why not).  

You can also visualize missing data (e.g., using a heatmap) to identify
patterns in missingness and determine if it is critical. 

### Conduct Bivariate Analysis  

Calculate covariance and correlation 

Utilize Pearson correlation for continuous/ values.  

For categorical variables, you could explore additional methods like
Spearman rank correlation (Optional)  

Create visualization 

You may visualize bivariate relationships with scatter plots, pair
plots, or heatmaps for correlation matrices 

You can create advanced visualization with Seaborn (Optional)  

### Test Statistics  

Develop a hypothesis and select the appropriate test statistics based on
your data. 

For each test (e.g., t-test, F-test, Chi-squared), be sure to discuss
the assumptions and whether they hold for your dataset 

Explain the test statistics results, interpret p-values and draw
conclusion  

### Regression and Classification 

Regression Analysis 

Select an appropriate regression model based on the variable selected.  

Run the regression model and evaluate it with the performance metrics

Interpret your result 

You may also want to check for multicollinearity or analyze residual to
check your assumptions (Optional) 

Classification Model  

Select an appropriate classification model based on the variable

Conduct the classification analysis and evaluate it with multiple

Interpret your result

Clustering Analysis

Evaluate the model performance metrics and interpret the results
